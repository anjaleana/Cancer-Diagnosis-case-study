### Model Training (Finding a problem Solution)

```python
X = df_cancer.drop(['target'],axis = 1) # axis means = 1 column wise drop. We dropped the target
X
```

```python
y = df_cancer['target']
y
```

Observation: We found the X and y dataset. Now We have to train the model(that's means we need to get function in which we can test our test data set)

#### We need to split the data set into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=5)
```

We are going to print the X_train and y_train 

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

We gonna use to SVM model

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import confusion_matrix
```

```python
svc_model = SVC()
```

```python
import warnings
warnings.filterwarnings("ignore")
svc_model.fit(X_train,y_train)
```

### Step 5 : Evaluting the Model

```python
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cm,annot = True)
```

The  TNR  value is 0 and TPR rate is 66 But FPR is 48 it is creating the problem

### Step 6 :Improving the Model

We will do normailization . Feature scaling(Unity-based Normalization) brings all values into range[0,1]

```python
min_train = X_train.min()
```

```python
max_train = X_train.max()

X_range = max_train-min_train
```

```python
X_train_scaled = (X_train-min_train)/X_range
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x=X_train['mean area'],y = X_train['mean smoothness'],hue= y_train)
```

#### Observation : This plot is between mean area  and mean smoothness without unit normalization

```python
sns.scatterplot(x=X_train_scaled['mean area'],y = X_train_scaled['mean smoothness'],hue=y_train)
```

Observation : The visualization of point is the same without normalizationBut the range of mean area and mean smoothness is between[0,1]

```python
min_test = X_test.min()
max_test = X_test.max()
range_test = max_test - min_test
```

```python
X_test_scaled = (X_test-min_test)/range_test
```

```python
svc_model.fit(X_train_scaled,y_train)
```

```python
y_test_scaled_predict = svc_model.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test,y_test_scaled_predict)
```

```python
sns.heatmap(cm,annot=True)
```

Observation: Now, we got good model because TPR and TNR is high 

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_test_scaled_predict))
```

### Improving the Model Part 2

```python
param_grid = {'C':[0.1,1,10,100],'gamma': [1,0.1,0.01,0.001],'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
```

```python
grid.fit(X_train_scaled,y_train)
```

```python
grid.best_params_
```

Observation: We got the best C is 10 and best gamma is 0.1

```python
grid_prediction = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test,grid_prediction)
```

```python
sns.heatmap(cm,annot = True)
```

Observation : We found the good model because the TPR and TNR is high and there are only 3 misclassified point

```python
print(classification_report(y_test,grid_prediction))
```

We found the 97% accuracy
